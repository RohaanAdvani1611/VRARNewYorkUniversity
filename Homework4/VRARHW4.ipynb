```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
```

```python
image = io.imread("https://www.printablee.com/postpic/2015/06/free-printable-eye-chart_112469.jpg")
image = cv2.resize(image, (360, 480))
cv2_imshow(image)
```

```python
f = float(input("Enter the focal length of the eye (in mm): "))
D = float(input("Enter the aperture (in mm): "))
S1 = float(input("Enter the focal distance of the image (in mm): "))
S = float(input("Enter the 'actual' distance from the image to the eye (in mm): "))
```

```python
M = f * 1. / (S1 - f)
c = M * D * abs(S - S1) * 1. / S
print(c)

blur_radius = c / 2
print(blur_radius)
```

```python
blurred_img = cv2.GaussianBlur(image, (0, 0), blur_radius)
cv2_imshow(blurred_img)
```
